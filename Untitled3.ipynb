```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta = '/content/drive/MyDrive/PM25_RUOA_vs_CAMS_formato_unico.csv'
df = pd.read_csv(ruta)
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
ruoa = df[df['fuente'] == 'RUOA'].copy()
cams = df[df['fuente'] == 'CAMS'].copy()
```

```python
comparado = pd.merge(ruoa[['fecha', 'PM25']], cams[['fecha', 'PM25']], on='fecha', suffixes=('_ruoa', '_cams'))
comparado['diferencia'] = comparado['PM25_cams'] - comparado['PM25_ruoa']
```

```python
plt.figure(figsize=(12,5))
plt.plot(comparado['fecha'], comparado['PM25_ruoa'], marker='o', label='RUOA')
plt.plot(comparado['fecha'], comparado['PM25_cams'], marker='s', label='CAMS (Satélite)')
plt.title('PM2.5 mensual: RUOA vs CAMS')
plt.xlabel('Fecha')
plt.ylabel('PM2.5 (µg/m³)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print(comparado[comparado['fecha'].dt.year == 2023])
```

```python
corr = comparado['PM25_ruoa'].corr(comparado['PM25_cams'])
mae = (comparado['diferencia']).abs().mean()
rmse = ((comparado['diferencia'])**2).mean()**0.5

print(f" Correlación: {corr:.2f}")
print(f" MAE (Error absoluto medio): {mae:.2f} µg/m³")
print(f" RMSE (Error cuadrático medio): {rmse:.2f} µg/m³")
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=comparado, y='diferencia')
plt.title('Distribución de diferencia (CAMS - RUOA)')
plt.ylabel('Diferencia PM2.5 (µg/m³)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print(f"\n Correlación (r): {corr:.2f}")

print(f" MAE (Error absoluto medio): {mae:.2f} µg/m³")
print(f" RMSE (Error cuadrático medio): {rmse:.2f} µg/m³")
```

```python
comparado['año'] = comparado['fecha'].dt.year
error_anual = comparado.groupby('año')['diferencia'].agg(['mean', 'std', 'count']).reset_index()
print("\n Error promedio por año:")
print(error_anual)
```
